DATA VISUALIZATION

```python
#HISTOGRAM(Sales)
import matplotlib.pyplot as plt
plt.hist(df['Sales'])
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
#BAR CHART(Category)
df['Category'].value_counts().plot(kind='bar')
plt.title('Sales by Category')
plt.show()
```

```python
df['Region'].value_counts().plot(kind='bar')
plt.title('Sales by Region')
plt.show()
```

```python
#Top 5 Products by Sales
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5).plot(kind='bar')
plt.title('Top 5 Products by Sales')
plt.show()
```

```python
df.groupby('Region')['Sales'].sum().sort_values(ascending=False).head(5).plot(kind='bar')
plt.title('Top 5 Regions by Sales')
plt.show()
```

```python
df.groupby('Order Month')['Sales'].sum()
```

```python
df.groupby('Category')['Sales'].sum()
```

```python
df['Sales'].mean()
```

```python
df.groupby('Order Month')['Sales'].sum().plot()
plt.title("Monthly Sales Trend")
plt.show()
```

CORRELATION ANALYSIS

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.scatter(df['Order Month'], df['Sales'])
plt.title("Sales vs Month")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
sns.boxplot(x=df['Sales'])
plt.title("Sales Outliers")
plt.show()
```
